```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
origin = pd.read_csv("../dataset/results-2021-12-18.csv")
origin
```

```python
alg = origin["algorithm"]
alg_count = alg.value_counts()
```

```python
fig, ax = plt.subplots()
label = ax.barh(alg_count.index, alg_count.values)
ax.bar_label(label, alg_count.values)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_xlabel("Number of Records")
ax.set_ylabel("Algorithm");
```

```python
iter_num = origin["iteration"]
iter_num_count = iter_num.value_counts().sort_index()
iter_num_count
```

```python
fig, ax = plt.subplots()
ax.scatter(iter_num_count.index, iter_num_count.values)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_xlabel("Times of Iteration")
ax.set_ylabel("Number of Records");
```

```python
svm_data = origin[origin["algorithm"] == "SVM"]
svm_data
```

```python
fig, ax = plt.subplots()
ax.scatter(svm_data["no_features"], svm_data["train_energy(J)"])
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_xlabel("no_features")
ax.set_ylabel("train_energy(J)");
```

```python
fig, ax = plt.subplots()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_xlabel("no_features")
ax.set_ylabel("no_datapoints")
```

```python
def draw(painter, algorithm):
    data = origin[origin["algorithm"] == algorithm]
    ax.scatter(data["no_features"], data["no_datapoints"], s=data["train_energy(J)"] * 40, alpha=0.2, label=algorithm)
```

```python
for al in alg_count.index:
    draw(ax, al)
ax.legend()
fig
```

```python
def draw_feat(painter, algorithm):
    data = origin[origin["algorithm"] == algorithm]
    ax.scatter(data["no_features"], data["train_energy(J)"], label=algorithm, alpha=0.4)
```

```python
fig, ax = plt.subplots()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_xlabel("no_features")
ax.set_ylabel("train_energy(J)")
ax.set_yscale('log')
for al in alg_count.index:
    draw_feat(ax, al)
ax.legend()
```

```python
def draw_datap(painter, algorithm):
    data = origin[origin["algorithm"] == algorithm]
    ax.scatter(data["no_datapoints"], data["train_energy(J)"], label=algorithm, alpha=0.2)
```

```python
fig, ax = plt.subplots()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_xlabel("no_datapoints")
ax.set_ylabel("train_energy(J)")
ax.set_yscale('log')
for al in alg_count.index:
    draw_datap(ax, al)
ax.legend()
```
